**Solución a la ecuación de Dirichlet b).**

La solución de la ecuación:
$$\frac{∂^2 u}{∂x^2} + \frac{∂^2u}{∂y^2} = 0, \hspace{1.5cm} 0<x<1, \hspace{1.5cm} 0<y<1$$
Con condiciones de frontera: 
$$b)\hspace{0.5cm} u(0,y) = 10y; \hspace{1cm} \frac{\partial u(1,y)}{∂x} = 1;$$ 
$$u(x,0) \hspace{0.5cm}= \hspace{0.5cm} u(x,1)\hspace{0.5cm}=\hspace{0.5cm}200$$

Está dada en términos de la serie de Fourier por la expresión:
$$u(x,y) = \sum_{n=1}^\infty (A_n cosh (n\pi x)+ B_n sinh(n\pi x))sin(n\pi y),$$
donde:
$$A_n = \frac{20(-1)^n}{n\pi}$$
$$B_n = \frac{2[(-1)^n-1]}{(n\pi)^2 cosh(n\pi)}-\frac{20tanh(n\pi)(-1)^{n+1}}{n\pi}$$

```python
from pylab import *
x = linspace(0,1.0,1000)
y = linspace(0,1.0,1000)

def u_x_y(x,y):
    N = 10
    f = zeros((len(x),len(y)))
    for i in range(len(y)):
        for n in range(1,N):
          A = (20*(-1)^(n+1))/(n*pi)
          C = (20*tanh(n*pi)*((-1)^(n+1)))/(n*pi)
          D = (2*( (-1)^(n) -1))
          G = (n*pi)**(2)
          F= ((G)*cosh(n*pi))
          B = D/F - C
          f[:,i] += (A*cosh(n*pi*x[i]) + B*sinh(n*pi*x[i]))*sin(n*pi*y) 
    return f


 
#%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12,10))
ax = fig.add_subplot(111, projection="3d")

X,Y = np.meshgrid(x,y)

U = u_x_y(x,y)

plt.xlabel('$y$'); plt.ylabel('$x$'); 
plt.title("u(x,y)")
ax.plot_surface(Y, X, U, cmap="gist_heat");
```
